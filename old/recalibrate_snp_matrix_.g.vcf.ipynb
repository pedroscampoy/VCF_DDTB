```python
def recheck_variant_gvcf(position, sample, gvcf_folder):
    for root, _, files in os.walk(gvcf_folder):
        for name in files:
            filename = os.path.join(root, name)
            if name.startswith(sample) and name.endswith("g.vcf"):
                with open(filename, 'r') as f:
                    content = f.read()
                    position_tab = "\t" + str(position) + "\t"
                    return int(position_tab in content)
```

```python
def variant_is_present(vcf_df,position):
    if position in vcf_df.POS.tolist():
        index_position = vcf_df.index[vcf_df.POS == position][0]
        if vcf_df.loc[index_position,'ALT_AD'] > vcf_df.loc[index_position,'REF_AD']:
            return "1"
        else:
            return "0"
    else:
        return "0"
```
